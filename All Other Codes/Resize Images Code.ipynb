# Accepted Crop Image Code 

```python
from PIL import Image
import os

def crop_and_save(file_path, output_folder, crop_box):
    # Open the image
    image = Image.open(file_path)
    
    # Crop the image
    cropped_image = image.crop(crop_box)
    
    # Save the cropped image
    output_path = os.path.join(output_folder, os.path.basename(file_path))
    cropped_image.save(output_path)

def batch_crop_and_save(input_folder, output_folder, crop_box):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate through all files in the input folder
    for file_name in os.listdir(input_folder):
        file_path = os.path.join(input_folder, file_name)
        
        # Check if the file is an image
        if os.path.isfile(file_path) and file_path.lower().endswith(('.jpg', '.jpeg', '.png')):
            crop_and_save(file_path, output_folder, crop_box)

# Example usage
input_folder = 'C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Subject5'
output_folder = 'C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Subject_5'
crop_box = (270, 250, 650, 500)  # Specify the crop box coordinates (left, upper, right, lower)

batch_crop_and_save(input_folder, output_folder, crop_box)

print('Process Completed Successfully')
```

# Select Random Image from different folder

```python
import os
import random
import shutil
```

```python

# Define the source folders where the images are located
source_folders = ['Train/Subject1/', 
                  'Train/Subject2/', 
                  'Train/Subject3/',
                  'Train/Subject4/',
                  'Train/Subject5/'
                 ]

# Define the destination folder where the selected images will be saved
destination_folder = 'Test/'

# Define the number of random images to select from each folder
num_images_to_select = 600

# Create the destination folder if it doesn't exist
#if not os.path.exists(destination_folder):
    #os.makedirs(destination_folder)

# Iterate over each source folder
for folder in source_folders:
    # Get the list of files in the folder
    files = os.listdir(folder)
    
    # Select random images from the list
    selected_images = random.sample(files, num_images_to_select)
    
    # Copy the selected images to the destination folder
    for image in selected_images:
        source_path = os.path.join(folder, image)
        destination_path = os.path.join(destination_folder, image)
        shutil.copy(source_path, destination_path)
        print(f"Image '{image}' copied to {destination_folder}")

print("Random image selection and copying completed.")
```

```python
# -------------------------------------------------------------------------------------------------------
# KC: To prepare dataset including train_set, validation_set, and test_set
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# KC: ImageDataGenerator can create an object used to generate batches of image data
image_datagen = ImageDataGenerator(rescale = 1./255, dtype='float32', validation_split=0.2)

# KC: The ImageDataGenerator class has the method flow_from_directory() 
#       wihch can read the images from folders containing images.
train_set = image_datagen.flow_from_directory(  'Train/',
                                                 target_size = (380, 250),
                                                 batch_size  = 10,
                                                 class_mode  = 'categorical',
                                                 subset='training'
                                                 )

validation_set = image_datagen.flow_from_directory(  'Train/',
                                                 target_size = (380, 250),
                                                 batch_size  = 10,
                                                 class_mode  = 'categorical',
                                                 subset='validation'
                                                 )
"""
test_set = image_datagen.flow_from_directory('final_project_dataset/Test/',
                                                target_size = (28, 28),
                                                batch_size  = 10,
                                                class_mode  = 'categorical',
                                                )
"""
```

```python
from keras import models
from keras import layers
from keras import losses
#import tensorflow as tf
# KC: to build up a model
# ------------------------------------------------[ Densely Connected Model ] -----------------------------
# network = models.Sequential()
# network.add(layers.Flatten(input_shape=(28,28,3)))
# #flatten
# network.add(layers.Dense(128, activation='relu'))
# network.add(layers.Dense(10, activation='softmax'))

network = models.Sequential([
    layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', input_shape=((380, 250,3)),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu'),
    layers.AveragePooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dense(units=64, activation='tanh'),
    layers.Dense(units=64, activation='tanh'),
    layers.Dense(units=128, activation='tanh'),
    layers.Dense(units=10, activation='softmax')
])

# Compile the model
#network.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
network.summary()

network.compile(optimizer='adam', 
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# KC: start to train the model
history = network.fit(  train_set,
                        validation_data  = validation_set,
                        epochs           = 15 )

print("Train accuracy : ", str(history.history['accuracy'][-1]))
print("Val accuracy : ", str(history.history['val_accuracy'][-1]))

print("Train loss : ", str(history.history['loss'][-1]))
print("Val loss : ", str(history.history['val_loss'][-1]))


# KC: To show the evaluation resuls of the Test set
test_loss, test_acc = network.evaluate(test_set)
print(' ')
print('Accuracy for Testset:', test_acc)
print(' ')
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the pre-trained model
# model = tf.keras.models.load_model('your_model_path')

# Specify the layer index for feature map visualization
layer_index = 3  # Adjust this index according to the layer you want to visualize

# Create a new model that outputs the desired layer activations
feature_extractor = tf.keras.Model(inputs=network.inputs, outputs=network.layers[layer_index].output)

# Load and preprocess an example image
image = tf.keras.preprocessing.image.load_img(r"C:\Users\SHAKIB\Documents\AI Projects\DL Final Project\final_project_dataset\Train\8\17.jpg", target_size=(28, 28))
image_array = tf.keras.preprocessing.image.img_to_array(image)
image_array = tf.expand_dims(image_array, axis=0)
image_array = tf.keras.applications.vgg16.preprocess_input(image_array)  # Adjust preprocessing for your model

# Extract the feature map activations for the input image
feature_maps = feature_extractor.predict(image_array)

# Visualize the feature maps
fig, axs = plt.subplots(nrows=8, ncols=8, figsize=(12, 12))
for i, ax in enumerate(axs.flat):
    ax.imshow(feature_maps[0, :, :, i], cmap='gray')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

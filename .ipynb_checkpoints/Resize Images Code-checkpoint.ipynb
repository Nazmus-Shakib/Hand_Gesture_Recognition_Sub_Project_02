```python
#empty lists
image_list = []
resized_images = []
```

```python
#append images to list
for filename in glob.glob('C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Subject1/*.png'):
    print(filename)
    img = Image.open(filename)
    image_list.append(img)
```

```python
#append resized images to list
for image in image_list:
    #image.show()
    image = image.resize((188, 90))
    resized_images.append(image)
```

```python
#save resized images to new folder
# could also probably use os to create new folder
for (i, new) in enumerate(resized_images):
    new.save('{}{}{}'.format('C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Resized_Subject1/color',i+1,'.png'))
```

# Accepted Crop Image Code 

```python
from PIL import Image
import os

def crop_and_save(file_path, output_folder, crop_box):
    # Open the image
    image = Image.open(file_path)
    
    # Crop the image
    cropped_image = image.crop(crop_box)
    
    # Save the cropped image
    output_path = os.path.join(output_folder, os.path.basename(file_path))
    cropped_image.save(output_path)

def batch_crop_and_save(input_folder, output_folder, crop_box):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate through all files in the input folder
    for file_name in os.listdir(input_folder):
        file_path = os.path.join(input_folder, file_name)
        
        # Check if the file is an image
        if os.path.isfile(file_path) and file_path.lower().endswith(('.jpg', '.jpeg', '.png')):
            crop_and_save(file_path, output_folder, crop_box)

# Example usage
input_folder = 'C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Subject1'
output_folder = 'C:/Users/SHAKIB/Documents/AI Projects/SubProject02/Resized_Subject1'
crop_box = (370, 250, 550, 350)  # Specify the crop box coordinates (left, upper, right, lower)

batch_crop_and_save(input_folder, output_folder, crop_box)
```
